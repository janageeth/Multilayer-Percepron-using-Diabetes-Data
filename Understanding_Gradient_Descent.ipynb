## **Understanding Gradient Descent in Linear Regression**



The objective in linear regression is to find the values of parameters m and b that minimize the cost function, often represented as the mean squared error (MSE) or a similar function. The cost function J(m,b) is defined as:

cost function.png

Here, n is the number of data points, xi and yi are the features and labels of the i-th data point, and m and b are the parameters we want to optimize.

Gradient descent is an iterative optimization algorithm that aims to minimize the cost function. The update rules for m and b in each iteration are based on the partial derivatives of the cost function with respect to m and b, denoted as ∂J/∂m and ∂b/∂J. These derivatives are calculated as:

Derivatives.png

The update steps for *m* and *b* in each iteration are given by:

Updates.png

These steps are repeated for a specified number of iterations until convergence.

```python
def gradient_descent(x, y):
  m_curr = b_curr = 0 #initialisation
  max_iterations = 1000
  learning_rate = 0.001
  n = len(x)
  for i in range(max_iterations):
    y_predicted = m_curr * x + b_curr
    md = -(2/n) * sum(x*(y -y_predicted))
    bd = -(2/n) * sum(y_predicted)
    m_curr = m_curr - learning_rate * md
    b_curr = b_curr - learning_rate * bd
    print("m {}, b {}, iteration {}".format(m_curr,b_curr,i))
```

```python
import numpy as np


x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

gradient_descent(x, y)
```

```python
import numpy as np

def gradient_descent(x, y):
    m_curr = b_curr = 0  # initialization
    max_iterations = 1000
    learning_rate = 0.001
    n = len(x)

    # Lists to store the cost and iteration values for plotting
    costs = []
    iterations = []

    for i in range(max_iterations):
        y_predicted = m_curr * x + b_curr

        # Calculate the mean squared error (MSE) as the cost
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        costs.append(cost)
        iterations.append(i)

        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)

        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        print("m {}, b {}, cost {}, iteration {}".format(m_curr, b_curr, cost, i))
```

```python
import numpy as np


x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

gradient_descent(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0  # initialization
    max_iterations = 1000
    learning_rate = 0.001
    n = len(x)

    # Lists to store the cost and iteration values for plotting
    costs = []
    iterations = []

    for i in range(max_iterations):
        y_predicted = m_curr * x + b_curr

        # Calculate the mean squared error (MSE) as the cost
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        costs.append(cost)
        iterations.append(i)

        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)

        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd

        print("m {}, b {}, cost {}, iteration {}".format(m_curr, b_curr, cost, i))

    # Plot the cost over iterations
    plt.plot(iterations, costs)
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Gradient Descent Convergence')
    plt.show()
```
